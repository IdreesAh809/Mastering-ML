**Visualize the Distribution of the Target Variable**

```python
# Count plot for the target variable
sns.countplot(x=df['num'])
plt.title("Distribution of Heart Disease Cases")
plt.xlabel("Heart Disease Presence (0 = No, 1+ = Yes)")
plt.ylabel("Count")
plt.show()
```
